```python
def converter_formato_imagens(pasta_entrada, pasta_saida, novo_formato):

    if not os.path.exists(pasta_saida):
        os.makedirs(pasta_saida)

    arquivos = os.listdir(pasta_entrada)

    for arquivo in arquivos:
    
            caminho_entrada = os.path.join(pasta_entrada, arquivo)
            
            nome_saida = os.path.splitext(arquivo)[0] + '.' + novo_formato.lower()
            caminho_saida = os.path.join(pasta_saida, nome_saida)
            if os.path.exists(caminho_saida):
                 print("Imagem ja existe")
                 continue

            try:
                 
                with Image.open(caminho_entrada) as imagem:
                    imagem.save(caminho_saida, format=novo_formato.upper())

                    print('imagem convertida', caminho_saida)
            except UnidentifiedImageError or KeyError or OSError:
                 print("Imagem com formato não suportado")
                 continue
# Exemplo de uso
pasta_entrada = '/Users/lucas/Developer/tcc/datasets/full/images_png'
pasta_saida = '/Users/lucas/Developer/tcc/datasets/full/images_jpg'
novo_formato = "JPEG"

converter_formato_imagens(pasta_entrada, pasta_saida, novo_formato)
```
