```python
def alter_extension(folder, extension, to_extension):
    png_files = glob.glob(os.path.join(folder, f'*.{extension}'))

    for file in png_files:
        jpg_file = os.path.splitext(file)[0] + f'.{to_extension}'

        os.rename(file, jpg_file)

        print(f"Renomeado: {os.path.basename(file)} para {os.path.basename(jpg_file)}")

    
```
